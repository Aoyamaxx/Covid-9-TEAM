```python
import pandas as pd
from datetime import datetime
```

```python
df1 = pd.read_csv('processed_data/unemployment_states.csv')
```

```python
df1['target_end_date'] = pd.to_datetime(df1['Label'], format='%Y %b')
```

```python
df1
```

```python
df2 = pd.read_csv("week5_data/weekly_aggregated_sup_data.csv")
```

```python
#get month
df1['year_month'] = df1['target_end_date'].dt.to_period('M')

# get month
df2['target_end_date'] = pd.to_datetime(df2['target_end_date'])
df2['year_month'] = df2['target_end_date'].dt.to_period('M')

# merge on location_key and year_month
merged_df = pd.merge(df2, df1, on=['location_key', 'year_month'], how='left')

merged_df.drop(columns=['target_end_date_y', 'Year', 'Period', 'Label', 'Series ID', 'year_month'], inplace=True)

merged_df.rename(columns={'target_end_date_x': 'target_end_date', 'Value': 'unemployment_rate'}, inplace=True)

merged_df
```
